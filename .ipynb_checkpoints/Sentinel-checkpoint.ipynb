```python
import datetime
import numpy as np

import matplotlib.pyplot as plt
```

```python
from sentinelhub import WmsRequest, WcsRequest, MimeType, CRS, BBox, CustomUrlParam
```

```python
def plot_image(image, factor=1):
    """
    Utility function for plotting RGB images.
    """
    fig = plt.subplots(nrows=1, ncols=1, figsize=(15, 7))

    if np.issubdtype(image.dtype, np.floating):
        plt.imshow(np.minimum(image * factor, 1))
    else:
        plt.imshow(image)
```

```python
box = [-75.64298408105815, -14.20062882414364, -75.6396578867764, -14.197876401443217]
date = "2016-07-15"
polygon = "POLYGON((-75.6396578867764 -14.199701335951724, -75.64231663022875 -14.20062882414364, -75.64298408105815 -14.198802571721302, -75.64032613621433 -14.197876401443217, -75.6396578867764 -14.199701335951724))"
                    
#[[-75.6396578867764, -14.199701335951724],
# [-75.64231663022875, -14.20062882414364],
# [-75.64298408105815, -14.198802571721302],
# [-75.64032613621433, -14.197876401443217],
# [-75.6396578867764, -14.199701335951724]]
```

```python
import sentinelhub
NDVI_request = WmsRequest(
    layer='NDVI',
    bbox=BBox(bbox=box, crs=CRS.WGS84),
    time="2020-04-20",
    width=500,
    image_format = MimeType.TIFF_d32f,    
    custom_url_params={
                        CustomUrlParam.GEOMETRY: polygon
                        },
    config=config
    )
NDVIs = NDVI_request.get_data()

plot_image(NDVIs[0])
```

```python
def save_image(image, path, factor=1):
    fig = plt.subplots(nrows=1, ncols=1, figsize=(15, 7))

    if np.issubdtype(image.dtype, np.floating):        
        plt.imshow(np.minimum(image * factor, 1))
    else:
        plt.imshow(image)
    plt.savefig(path)
    plt.close()
    
save_image(NDVIs[0], "outputs\\images\\image1.png")       
```

```python
import json
from pandas import json_normalize
with open('inputs\\farm_map.json') as f:
    data = json.load(f)
df_data = json_normalize(data["features"]) 
df_data
```

```python
df_data["Kg/Ha"] = df_data["properties.values"].apply(lambda x: x[0]["value"])

def poly_to_box(x):
    lon = []
    lat = []
    for i in x[0]:
        lon += [i[0]]
        lat += [i[1]]
    
    box = [max(lon), max(lat), min(lon), min(lat)]
    return(box)  

df_data["bbox_coordinates"] = df_data["geometry.coordinates"].apply(lambda x: poly_to_box(x))
df_data[["Kg/Ha", "bbox_coordinates"]]
```
